```python
mod1 = sm.ols(formula = mod1_formula, data = dt)
mod1_res = mod1.fit()
print(mod1_res.summary())
```

```python
mod2 = sm.ols(formula = mod2_formula, data = dt)
mod2_res = mod2.fit()
print(mod2_res.summary())
```

```python
prstd, ln_iv_l, ln_iv_u = wls_prediction_std(mod2_res)
```

```python
iv_l = np.exp(ln_iv_l)
iv_u = np.exp(ln_iv_u)
# predicted = np.exp()
```

```python
error_margin = np.mean((iv_u - iv_l))
```
